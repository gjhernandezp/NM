```
import mglearn
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
model = KNeighborsRegressor().fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model.predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()

```

```
import mglearn
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
model = MLPRegressor(random_state=1, max_iter=500).fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model .predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()
```

```
import mglearn
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=40)
model = MLPRegressor(hidden_layer_sizes=(100,50,20),random_state=1, max_iter=500, activation='relu').fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model .predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()
```
