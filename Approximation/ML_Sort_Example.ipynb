# Learning to sort with a Neural Network (MLPRegressor) 

```
import random as rnd
import numpy as np

n=10
Nexamples=1000

X =[]
y =[]
for j in range(Nexamples):
    y.append(np.array(range(n)))
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))
```

```
import pandas as pd
data = pd.DataFrame({'X': [a for a in X], 'y':[a for a in y]})
data
```

```
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(100,50,20),random_state=1, max_iter=1000, activation='logistic').fit(X,y)
y_pred =model.predict(X)

import matplotlib.pyplot as plt
plt.plot(model.loss_curve_)
plt.show()
```

```
import pandas as pd
y_p = pd.DataFrame({'y_pred':[ [round(b,3) for b in a] for a in y_pred]})
y_p 
```

```
import random as rnd
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import pandas as pd

n=10
Nexamples=1000

X =[]
y =[]

for j in range(Nexamples):
    y.append(np.array(range(n)))
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))

for j in range(Nexamples):
    y.append(2*np.array(range(n)))
    list1 = list(2*np.array(range(n)))
    rnd.shuffle(list1)
    X.append(np.array(list1))

model = MLPRegressor(hidden_layer_sizes=(100,50,20), max_iter=1000, activation='tanh').fit(X,y)
y_pred =model.predict(X)


plt.plot(model.loss_curve_)
plt.show()

print('Accuracy testing : {:.3f}'.format(model.score(X,y)))

data = pd.DataFrame({'X': [a for a in X], 'y':[a for a in y], 'y_pred':[ [int(round(b)) for b in a] for a in y_pred]})
data
```
