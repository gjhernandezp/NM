Sorting with a Neural Network (MLPRegressor) 

```
import random as rnd
import numpy as np

n=10
Nexamples=1000

X =[]
y =[]
for j in range(Nexamples):
    list1 = list(range(n))
    y.append(np.array(list(range(n))))
    rnd.shuffle(list1 )
    X.append(np.array(list1))
```

```
import pandas as pd
data = pd.DataFrame({'X': [a for a in X], 'y':[a for a in y]})
data
```

```
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(100,50,20),random_state=1, max_iter=500, activation='logistic').fit(X,y)
y_pred =model.predict(X)

import matplotlib.pyplot as plt
plt.plot(model.loss_curve_)
plt.show()
```

```
import pandas as pd
y_p = pd.DataFrame({'y_pred':[ [round(b,3) for b in a] for a in y_pred]})
y_p 
```

```
import pandas as pd
data = pd.DataFrame({'X': [a for a in X], 'y':[a for a in y],'y_pred':[ [int(round(b)) for b in a] for a in y_pred]})
data
```

```
n=10
Nexamples=100000

X_test =[]
for j in range(Nexamples):
    list1 = list(range(n))
    rnd.shuffle(list1 )
    X_test.append(np.array(list1))

y_test = model.predict(X_test)

print('Accuracy testing : {:.3f}'.format(model.score(X_test, y_test)))

import pandas as pd
data = pd.DataFrame({'Xtest': [a for a in X_test], 'y_test':[[int(round(b)) for b in a] for a in y_test]})
data
```

```
from itertools import permutations

n=10
a= range(n)
X_test =[]
for P in permutations(a):
  X_test.append(np.array(P))

y_test = model.predict(X_test)

print('Accuracy testing : {:.3f}'.format(model.score(X_test, y_test)))

import pandas as pd
data = pd.DataFrame({'Xtest': [a for a in X_test], 'y_test':[[int(round(b)) for b in a] for a in y_test]})
data
```
