````
eps = 1.0
n = 1
while eps != 0.0 :
  eps /= 2 
  n += 1 
  print(n,eps)
````

## Machine Infinity

````
import math
eps = 2.0
n = 1
while eps != math.inf :
  eps *= 2 
  n += 1 
  print(n,eps)
````
